```python
data_true_manual_testing = data_true.tail(10)
for i in range(21416,21406,-1):
    if i in data_true.index:
        data_true.drop([i], axis = 0, inplace = True)
```

```python
data_merge = pd.concat([data_fake, data_true], axis =0 )
data_merge.head(10)
```

```python
data = data_merge.drop(["title", "subject","date"], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(["index"], axis = 1, inplace = True)
data.head()
```

```python
def wordopt(t):
    tt = t.lower()
    t = re.sub('\[.*?\]', '', t)
    t = re.sub("\\W"," ",t)
    t = re.sub('https?://\S+|www\.\S+', '', t)
    t = re.sub('<.*?>+', '', t)
    t = re.sub('[%s]' % re.escape(string.punctuation), '', t)
    t = re.sub('\n', '', t)
    t = re.sub('\w*\d\w*', '', t)
    return t
```

```python
data["text"] = data["text"].apply(wordopt)
```

```python
x = data["text"]
y = data["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression()
logReg.fit(xv_train,y_train)
```

```python
pred_lr=logReg.predict(xv_test)
logReg.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DecTree = DecisionTreeClassifier()
DecTree.fit(xv_train, y_train)
```

```python
pred_dt = DecTree.predict(xv_test)
DecTree.score(xv_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)
```

```python
pred_rfc = RFC.predict(xv_test)
RFC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_logReg = logReg.predict(new_xv_test)
    pred_DecTree = DecTree.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nRFC Prediction: {}".format(output_label(pred_logReg[0]), output_label(pred_DecTree[0]), output_label(pred_RFC[0])))

news = str(input())
manual_testing(news)
```
